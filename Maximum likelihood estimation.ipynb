```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
alpha_h=6
alpha_t=4
```

```python
def likelihood(theta):
    return np.power(theta,alpha_h) * np.power((1-theta),alpha_t)
```

```python
theta_tilda=np.array(np.arange(0,1.01,0.01),dtype=float)
```

```python
def plotLikelihood(theta):
    x = theta
    #alpha_h=alphah
    #alpha_t=alphat
    N=alpha_h+alpha_t
    y = np.array(likelihood(x))
    mle=np.argmax((np.array(likelihood(x))))
    plt.title("Likelihood for N={}".format(N)) 
    plt.xlabel("x axis caption") 
    plt.ylabel("y axis caption") 
    plt.plot(x,y) 
    plt.plot([x[mle] ,x[mle]], [0, y[mle]])
    plt.show()
```

```python
plotLikelihood(theta_tilda)
```

```python
alpha_h=3
alpha_t=2
plotLikelihood(theta_tilda)
```
